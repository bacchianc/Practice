```python
#import necessary libraries

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('IndianHouses.csv')
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.shape
```

```python
df['Locality'].value_counts()
```

```python
df.drop('Locality',axis=1,inplace=True)
df.head()
```

```python
df.Furnishing.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Furnishing_Label'] = label_encoder.fit_transform(df['Furnishing'])
df['Furnishing_Label'].value_counts()
```

```python
df['Furnishing_simple']=df.Furnishing.map({'Semi-Furnished':1,'Unfurnished':0,'Furnished':2})
```

```python
df['Furnishing_simple'].value_counts()
```

```python
df['Furnishing']=df['Furnishing_simple']
```

```python
df.head()
```

```python
df.drop(['Furnishing_Label','Furnishing_simple'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.Status.value_counts()
```

```python
df['Status']=df['Status'].map({'Ready_to_move':1,'Almost_ready':0})
```

```python
df.head()
```

```python
df.Transaction.value_counts()
```

```python
df['Transaction']=df['Transaction'].map({'Resale':0,'New_Property':1})
```
